```python
import pandas as pd
```

Load the Data

```python
df = pd.read_csv('dataset/Student_performance_data.csv', encoding='latin1')
```

```python
df.head()
```

Clean the Dataset

```python
df.drop(['StudentID','Gender','ParentalEducation','ParentalSupport','GradeClass'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# First, create the TotalActivities column
df['TotalActivities'] = df['Extracurricular'] + df['Sports'] + df['Music'] + df['Volunteering']
```

```python
df.drop(['Extracurricular', 'Sports', 'Music', 'Volunteering'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop(['Tutoring'], axis=1, inplace=True)
```

```python
df.to_csv('dataset/cleaned_student_performance_data_cleaned.csv', index=False)
```
