```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/used_car_price_dataset_extended.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.drop(['color','service_history','accidents_reported','insurance_valid'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop(['owner_count'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['brand'], dtype=int)
df.head()
```

```python
df = pd.get_dummies(df, columns=['transmission'], dtype=int)
df.head()
```

```python
df = pd.get_dummies(df, columns=['fuel_type'], dtype=int)
df.head()
```

```python
df.to_csv("dataset/cleaned_used_car_price_dataset.csv", index=False)
```
